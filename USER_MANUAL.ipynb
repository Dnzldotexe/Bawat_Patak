    </div>
</div>

<h1 style="font-family: Verdana; background-color:#C0AFE2; padding: 20px; display: inline-block;">
    1.3 Solution üí°
</h1>

<div style="display: flex; text-align: left;">
    <div style="background-color:#DDD5F3; padding: 20px;">
        <p style="font-size:20px; font-family:verdana; line-height: 1.7em; margin: 0; padding-left: 20px;">
            By simply logging your everyday water consumption, you can monitor how much water you consume. You would be aware of how much water is or was wasted on things that you wouldn't know that is contributing to your water consumption.
        </p>
    </div>
</div>

<h1 style="font-family: Verdana; background-color:#C0AFE2; padding: 20px; display: inline-block;">
    1.4 Project Description üìú
</h1>

<div style="display: flex; text-align: left;">
    <div style="background-color:#DDD5F3; padding: 20px;">
        <p style="font-size:20px; font-family:verdana; line-height: 1.7em; margin: 0; padding-left: 20px;">
            Bawat Patak, a user-friendly web app designed to raise awareness about water conservation. It allows individuals to effortlessly track and monitor their water consumption, empowering them to make informed decisions and contribute to the collective effort of preserving this precious resource. Every user becomes an advocate for sustainable water practices, ensuring that "bawat patak" counts towards a more water-conscious future.
        </p>
    </div>
</div>
